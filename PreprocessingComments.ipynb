```python
Comments=[]

for comment in df1["comment_text"]:
  if(len(comment)==0):
   Comments.append("")
   # print(comment)
   df1.drop(comment)

  #print(detect(comment))
  #if(detect(comment)=='en'):
    #Comments.append(comment)
  try:
    if(detect(comment)=='en'):
      Comments.append(comment)
    else:
      Comments.append("")
  except Exception as e:
    print(comment)
    #df1.drop(index=str(comment), inplace=True)
    #print(df1['comment_text'].loc[comment])
    #print(df1.get_loc(comment))
    #print(df1['comment_text'].index(comment))
    Comments.append("")
    #print("Error during language detection:", e)
len(Comments)
```

```python

nlp = spacy.load('en_core_web_sm')
def clean(text):
  doc=nlp(text)
  noPuncText=""
  for token in doc:
    if not token.is_punct :
      noPuncText=noPuncText+" "+token.text
  lowerText=""
  for word in noPuncText:
    lowerText=lowerText+word.lower()
  #stop word remove
  doc = nlp(lowerText)

  filtered_words = [token.text for token in doc if not token.is_stop]

  new_string = ' '.join(filtered_words)
  # eeeemoji
  emoj = re.compile("["
        u"\U00002700-\U000027BF"  # Dingbats
        u"\U0001F600-\U0001F64F"  # Emoticons
        u"\U00002600-\U000026FF"  # Miscellaneous Symbols
        u"\U0001F300-\U0001F5FF"  # Miscellaneous Symbols And Pictographs
        u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
        u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
        u"\U0001F680-\U0001F6FF"  # Transport and Map Symbols
                      "]+", re.UNICODE)
  return re.sub(emoj, '', new_string)


clean("Hi this is Arya !ü§ì")
```

```python
CleanComments=[]
for comment in Comments:
  CleanComments.append(clean(comment))
print(len(CleanComments))

df1['Clean Comments']=CleanComments
```
